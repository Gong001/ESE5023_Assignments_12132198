```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df_input = pd.read_csv('2281305.csv')
```

```python
df_wind = pd.DataFrame(np.full((len(df_input),1),np.nan),index = df_input['DATE'],columns = ['speed rate'])
for i in range(len(df_wind)):
    df_wind.iloc[i,0] = int(df_input['WND'][i][8:12])
```

```python
arr_wind = df_wind['speed rate'].values
arr_wind[arr_wind==9999] = np.nan
df_wind['speed rate'] = arr_wind
```

```python
df_wind.index = pd.to_datetime(df_wind.index)
```

```python
def plot10year():
    df_wind.resample('m').mean()['speed rate'].plot(ylabel = 'wind speed (m/s)')
plot10year()
```
